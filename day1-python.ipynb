```python
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))

if (a > b) and (a > c):
    print(a)
elif (b > c):
    print(b)
elif (c > b):
    print(c)
else:
    print("All are equal")
```

```python
marks = int(input("Enter the marks: " ))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade c")
else:
    print("Fail")
```

```python
#continue statement : odd numbers
for i in range(11):
    if i%2==0:
        continue
    print(i)
```
